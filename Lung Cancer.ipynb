```python
data_age=data2['AGE']
q3=data_age.quantile(0.75)
q1=data_age.quantile(0.25)
iqr=q3-q1
lower_limit=q1-(1.5*iqr)
upper_limit=q3+(1.5*iqr)
```

```python
def minmax_norm(df):
    return (data3-data3.min())/(data3.max()-data3.min())
data4=minmax_norm(data3)
```

```python
data4.head()
```

```python
x=data4.drop('LUNG_CANCER',axis=1)
y=data4['LUNG_CANCER']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression(random_state=0)
model1.fit(x_train,y_train)
```

```python
print('Training accuracy',model1.score(x_train,y_train))
print('Test accuracy',model1.score(x_test,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
class_dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
class_dt.fit(x_train,y_train)
```

```python
print('Training accuracy',class_dt.score(x_train,y_train))
print('Test accuracy',class_dt.score(x_test,y_test))
```
